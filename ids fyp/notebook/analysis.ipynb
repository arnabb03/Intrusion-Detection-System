```python
import numpy as np
import matplotlib.pyplot as plt

def plot_per_column_distribution(data, n_graph_shown, n_graph_per_row):
    nunique = data.nunique()
    columns_to_plot = data.columns[(nunique > 1) & (nunique < 50)]
    data_to_plot = data[columns_to_plot]

    n_rows, n_cols = data_to_plot.shape
    
    n_graph_row = (n_cols + n_graph_per_row - 1) // n_graph_per_row 
    plt.figure(num=None, figsize=(6 * n_graph_per_row, 8 * n_graph_row), dpi=80, facecolor='w', edgecolor='k')

    for i, col_name in enumerate(columns_to_plot[:n_graph_shown]):
        plt.subplot(n_graph_row, n_graph_per_row, i + 1)
        column_data = data_to_plot[col_name]

        if not np.issubdtype(column_data.dtype, np.number):
            value_counts = column_data.value_counts()
            value_counts.plot.bar()
        else:
            column_data.hist()

        plt.ylabel('Counts')
        plt.xticks(rotation=90)
        plt.title(f'{col_name} (Column {i})')

    plt.tight_layout(pad=1.0, w_pad=1.0, h_pad=1.0)
    plt.show()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

data_encoded = result_data.copy()

le = LabelEncoder()

data_encoded['Label'] = le.fit_transform(result_data['Label'])

label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
label_mapping_clean = {k: int(v) for k, v in label_mapping.items()}

print("Label Mapping:")
print(label_mapping_clean)
# Display the modified DataFrame
print("\nEncoded DataFrame:")
print(data_encoded.head())
```

## Smote

```python
from imblearn.over_sampling import SMOTE
import pandas as pd
from collections import Counter


X = data_encoded.drop(columns=['Label'])  
y = data_encoded['Label']                 


print("Original label distribution:")
print(y.value_counts())


class_counts = Counter(y)
max_count = max(class_counts.values())


sampling_dict = {
    cls: int(max_count * 0.5) 
    for cls, count in class_counts.items() 
    if count < max_count
}

print("\nSampling strategy dict (class: target count):")
print(sampling_dict)


smote = SMOTE(sampling_strategy=sampling_dict, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)


balanced_data = pd.DataFrame(X_resampled, columns=X.columns)
balanced_data['Label'] = y_resampled


balanced_data['Label'] = balanced_data['Label'].map({v: k for k, v in label_mapping_clean.items()})


print("\nNew label distribution after SMOTE:")
print(balanced_data['Label'].value_counts())


balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)


print("\nBalanced data sample:")
print(balanced_data.head())
```

# Model Training

## 1. Decision Tree

### Hyperparameter tuning using optuna

```python
import optuna
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, train_test_split


def objective(trial):
    
    criterion = trial.suggest_categorical("criterion", ["gini", "entropy"])
    max_depth = trial.suggest_int("max_depth", 2, 30)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 20)
    max_features = trial.suggest_categorical("max_features", [None, "sqrt", "log2"])

    
    clf = DecisionTreeClassifier(
        criterion=criterion,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        random_state=42
    )

    
    score = cross_val_score(clf, X_train, y_train, cv=5, scoring="accuracy")
    return score.mean()
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=20)
```

```python
print("Best Trial:")
print(f"  Accuracy: {study.best_trial.value:.4f}")
print("  Best Hyperparameters:")
for key, value in study.best_trial.params.items():
    print(f"    {key}: {value}")
```

## 2. Knn

### Hyperparameter tuning

```python
import optuna
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

def objective(trial):
    
    n_neighbors = trial.suggest_int('n_neighbors', 1, 10)
    weights = trial.suggest_categorical('weights', ['uniform', 'distance'])
    p = trial.suggest_int('p', 1, 2)  # p=1 is Manhattan, p=2 is Euclidean
    
    knn = KNeighborsClassifier(
        n_neighbors=n_neighbors,
        weights=weights,
        p=p
    )
    
    score = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    return score.mean()
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20)
```

## 3 . Random Forest

### Hyperparameter tuning

```python
import optuna
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

def objective(trial):
    n_estimators = trial.suggest_categorical('n_estimators', [100, 150, 200])
    max_depth = trial.suggest_categorical('max_depth', [20, 30, None])
    min_samples_split = trial.suggest_categorical('min_samples_split', [2, 5, 10])
    min_samples_leaf = trial.suggest_categorical('min_samples_leaf', [1, 2, 4])
    max_features = trial.suggest_categorical('max_features', ['sqrt', 'log2'])

    clf = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        random_state=42,
        n_jobs=-1
    )

    score = cross_val_score(clf, X_train, y_train, cv=5, scoring='accuracy')
    return score.mean()
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20)
```

```python
import optuna
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score

def objective(trial):
    model = CatBoostClassifier(
        iterations=trial.suggest_categorical('iterations', [10,50,100, 200]),
        learning_rate=trial.suggest_float('learning_rate', 0.1, 0.3),
        depth=trial.suggest_int('depth', 4, 8),
        loss_function='MultiClass',
        eval_metric='Accuracy',
        verbose=0,
        random_state=42
    )

    score = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy')
    return score.mean()
```

## 5. Stacking Ensemble

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from catboost import CatBoostClassifier

dt_clf = DecisionTreeClassifier(criterion='entropy',max_depth=30,min_samples_split=14,min_samples_leaf=10,
                                max_features=None,
                                random_state=42)
catboost_clf = CatBoostClassifier(
    iterations=200,
    learning_rate=0.24,
    depth=8,
    loss_function='MultiClass',
    eval_metric='Accuracy',
    auto_class_weights='Balanced',
    verbose=0,
    random_state=42
)
knn_clf = KNeighborsClassifier(n_neighbors=5, weights='uniform', p=1)

meta_model =  RandomForestClassifier(
    n_estimators=150,
    max_depth=30,
    min_samples_split=10,
    min_samples_leaf=2,
    max_features='sqrt',
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)
stacked_model = StackingClassifier(
    estimators=[
        ('DecisionTree', dt_clf),
        ('CatBoostClassifier', catboost_clf),
        ('KNN', knn_clf)
    ],
    final_estimator=meta_model,
    cv=5
)

stacked_model.fit(X_train, y_train)
```

```python
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')

def print_all_accuracy_metrics(algorithm, name):
    y_train_pred = algorithm.predict(X_train)
    y_test_pred = algorithm.predict(X_test)

    acc_train = metrics.accuracy_score(y_train, y_train_pred)
    acc_test = metrics.accuracy_score(y_test, y_test_pred)
    print(f"{name} : Accuracy on training data: {acc_train:.3f}")
    print(f"{name} : Accuracy on test data: {acc_test:.3f}\n")

    f1_train = metrics.f1_score(y_train, y_train_pred, average='macro')
    f1_test = metrics.f1_score(y_test, y_test_pred, average='macro')
    print(f"{name} : F1 score on training data: {f1_train:.3f}")
    print(f"{name} : F1 score on test data: {f1_test:.3f}\n")

    recall_train = metrics.recall_score(y_train, y_train_pred, average='macro')
    recall_test = metrics.recall_score(y_test, y_test_pred, average='macro')
    print(f"{name} : Recall on training data: {recall_train:.3f}")
    print(f"{name} : Recall on test data: {recall_test:.3f}\n")

    
    precision_train = metrics.precision_score(y_train, y_train_pred, average='macro')
    precision_test = metrics.precision_score(y_test, y_test_pred, average='macro')
    print(f"{name} : Precision on training data: {precision_train:.3f}")
    print(f"{name} : Precision on test data: {precision_test:.3f}\n")

   

    print(f"{name} : Classification Report on test data:\n")
    print(metrics.classification_report(y_test, y_test_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')

def plot_metrics_and_confusion_matrix(algorithm, name):
    y_train_pred = algorithm.predict(X_train)
    y_test_pred = algorithm.predict(X_test)

    metrics_dict = {
        'Accuracy': [metrics.accuracy_score(y_train, y_train_pred), metrics.accuracy_score(y_test, y_test_pred)],
        'F1 Score': [metrics.f1_score(y_train, y_train_pred, average='macro'), metrics.f1_score(y_test, y_test_pred, average='macro')],
        'Recall': [metrics.recall_score(y_train, y_train_pred, average='macro'), metrics.recall_score(y_test, y_test_pred, average='macro')],
        'Precision': [metrics.precision_score(y_train, y_train_pred, average='macro'), metrics.precision_score(y_test, y_test_pred, average='macro')],
    }

    labels = ['Train', 'Test']
    x = range(len(metrics_dict))  

    train_scores = [v[0] for v in metrics_dict.values()]
    test_scores = [v[1] for v in metrics_dict.values()]

    width = 0.35  

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

    ax1.bar(x, train_scores, width=width, label='Train', color='skyblue')
    ax1.bar([i + width for i in x], test_scores, width=width, label='Test', color='salmon')
    ax1.set_xticks([i + width/2 for i in x])
    ax1.set_xticklabels(list(metrics_dict.keys()))
    ax1.set_ylim(0, 1.1)
    ax1.set_ylabel('Score')
    ax1.set_title(f'Performance Metrics for {name}')
    ax1.legend()

    
    cm = metrics.confusion_matrix(y_test, y_test_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax2)
    ax2.set_title(f'Confusion Matrix for {name} (Test Data)')
    ax2.set_xlabel('Predicted label')
    ax2.set_ylabel('True label')

    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model, model_name):
    y_pred = model.predict(X_test)
    return {
        'Model': model_name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, average='macro'),
        'Recall': recall_score(y_test, y_pred, average='macro'),
        'F1 Score': f1_score(y_test, y_pred, average='macro')
    }
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


results = []
results.append(evaluate_model(clf, 'Decision Tree'))
results.append(evaluate_model(knn, 'KNN'))
results.append(evaluate_model(rd, 'Random Forest'))
results.append(evaluate_model(cb, 'CatBoost'))
results.append(evaluate_model(finalmodel, 'Stacking Ensemble'))

results_df = pd.DataFrame(results)


results_df_percent = results_df.copy()
for col in ['Accuracy', 'Precision', 'Recall', 'F1 Score']:
    results_df_percent[col] = results_df_percent[col] * 100


model_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']


fig, axes = plt.subplots(1, 2, figsize=(16, 6))


axes[0].barh(results_df_percent['Model'], results_df_percent['Accuracy'], color=model_colors)
axes[0].set_title('Model Accuracy (%)')
axes[0].set_xlabel('Accuracy (%)')
axes[0].set_xlim(90,100)

for i, val in enumerate(results_df_percent['Accuracy']):
    axes[0].text(val + 0.5, i, f'{val:.2f}%', va='center')


bar_height = 0.2
index = np.arange(len(results_df_percent))

axes[1].barh(index - bar_height, results_df_percent['Precision'], height=bar_height, label='Precision', color='orange')
axes[1].barh(index, results_df_percent['Recall'], height=bar_height, label='Recall', color='lightgreen')
axes[1].barh(index + bar_height, results_df_percent['F1 Score'], height=bar_height, label='F1 Score', color='lightcoral')

axes[1].set_yticks(index)
axes[1].set_yticklabels(results_df_percent['Model'])
axes[1].set_title('Precision, Recall, F1 Score (%)')
axes[1].set_xlabel('Score (%)')
axes[1].set_xlim(90, 100)
axes[1].legend()


for i in range(len(results_df_percent)):
    axes[1].text(results_df_percent['Precision'][i] + 0.5, i - bar_height, f"{results_df_percent['Precision'][i]:.2f}%", va='center')
    axes[1].text(results_df_percent['Recall'][i] + 0.5, i, f"{results_df_percent['Recall'][i]:.2f}%", va='center')
    axes[1].text(results_df_percent['F1 Score'][i] + 0.5, i + bar_height, f"{results_df_percent['F1 Score'][i]:.2f}%", va='center')

plt.tight_layout()
plt.savefig("comparison.png", dpi=300, bbox_inches='tight')
plt.show()
```
